## Scatterplot with Trendline (3 points)

```python
# Create a scatterplot with a trendline from two lists of values

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

num_points = 50
x = 5 + np.arange(num_points) + np.random.randn(num_points)
y = 10 + np.arange(num_points) + 5 * np.random.randn(num_points)

# If you want them combined into a dataframe
df = pd.DataFrame({'x': x, 'y': y})
```

```python
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
```

```python
plt.plot(x, p(x), color='red', label='Trend line')
plt.plot(x,y)
```

The error here was that my trend line wasn't linear. To make it linear, I created a new variable z which is built off of the data from x and y using the first degree polynomial (i.e. y = nx where n is in the interval (-inf, inf).) From there, I took z and plugged into a new plot called poly1d which plots a first degree polynomial. After that, I generated my standard x, y plot and then on top of it did a plot that used x and p which is the poly1d command which I populated into my variable.
